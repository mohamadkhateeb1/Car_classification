**2-استيراد المكتبات**

```python
# استيراد المكتبات
import os
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.model_selection import train_test_split
```

**3-تقسيم البيانات**


```python
# إعداد المسار إلى مجلد البيانات
data_dir = '/content/drive/MyDrive/Cars_Dataset'

# توليد البيانات مع التقسيم للتدريب والاختبار
train_datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2  # نسبة 20% للاختبار
)

# تحميل بيانات التدريب
train_generator = train_datagen.flow_from_directory(
    data_dir,
    target_size=(128, 128),  # حجم الصور (تقدر تغيره حسب احتياجك)
    batch_size=32,
    class_mode='categorical',  # لأن عندك تصنيف متعدد
    subset='training',
    shuffle=True
)

# تحميل بيانات الاختبار
val_generator = train_datagen.flow_from_directory(
    data_dir,
    target_size=(128, 128),
    batch_size=32,
    class_mode='categorical',
    subset='validation',
    shuffle=True
)
```

**4-بناء النموذج**

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout

# عدد الفئات (عدد مجلدات التصنيف)
num_classes = 7

# بناء النموذج
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 3)),
    MaxPooling2D(pool_size=(2, 2)),

    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),

    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),

    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(num_classes, activation='softmax')  # الطبقة النهائية للتصنيف
])
# ملخص النموذج
model.summary()
# تجميع النموذج
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',  # لأن التصنيف متعدد
    metrics=['accuracy']
)

```

**5-تدريب النموذج**

```python
# عدد الإيبوكس (مرات التكرار على البيانات) — ممكن تعدله حسب الأداء
epochs = 10

# تدريب النموذج
history = model.fit(
    train_generator,
    epochs=epochs,
    validation_data=val_generator
)
```

**6-رسم دقة التدريب**

```python
# رسم دقة التدريب والتحقق
plt.plot(history.history['accuracy'], label='دقة التدريب')
plt.plot(history.history['val_accuracy'], label='دقة التحقق')
plt.xlabel('Epoch')
plt.ylabel('الدقة')
plt.legend()
plt.title('دقة التدريب والتحقق')
plt.show()

# رسم الخسارة
plt.plot(history.history['loss'], label='خسارة التدريب')
plt.plot(history.history['val_loss'], label='خسارة التحقق')
plt.xlabel('Epoch')
plt.ylabel('الخسارة')
plt.legend()
plt.title('خسارة التدريب والتحقق')
plt.show()
```

**7-اختبار النموذج**

```python
# تقييم النموذج على بيانات التحقق
loss, accuracy = model.evaluate(val_generator)

print(f"دقة النموذج على بيانات الاختبار: {accuracy * 100:.2f}%")
print(f"قيمة الخسارة: {loss:.4f}")
```

**8-رسم دقة الاختبار**

```python
# استخراج الصور والتسميات الحقيقية من الـ generator
x_test, y_test = next(val_generator)

# توقع النموذج
preds = model.predict(x_test)
pred_labels = np.argmax(preds, axis=1)
true_labels = np.argmax(y_test, axis=1)

# أسماء الفئات
class_names = list(val_generator.class_indices.keys())

# عرض بعض النتائج
import random

plt.figure(figsize=(12, 8))
for i in range(9):
    idx = random.randint(0, len(x_test) - 1)
    plt.subplot(3, 3, i + 1)
    plt.imshow(x_test[idx])
    plt.axis('off')
    plt.title(f"TRUE: {class_names[true_labels[idx]]}\n anticipation: {class_names[pred_labels[idx]]}")
plt.tight_layout()
plt.show()
```
